```python
df = pd.read_csv(r"D:\Projects\YELP\data\friends.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
for i in range(0, df.shape[0], 300000):
    subset = df.iloc[i:i+300000]

    df_list = []
    for index, row in subset.iterrows():
        if index%50000 == 0:
            print('Completed ', index)
        user_id = row['user_id']
        friends = row['friends']
        friends = friends.split(',')
        friends = [x.strip() for x in friends]
        subsetNew = pd.DataFrame()
        subsetNew['friend_id'] = friends
        subsetNew['user_id'] = user_id
        df_list.append(subsetNew)
    masterSub = pd.concat(df_list, axis=0)
    savePath = r"D:\Projects\YELP\friends_{}.csv".format(str(i))
    masterSub.to_csv(savePath, index=False)
```

```python
master = pd.concat(df_list, axis=0)
master = master[['user_id', 'friend_id']]
```
